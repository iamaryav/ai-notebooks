```python
# Make a bigram model manually
count = {}
words[:1]
for w in words:
    nw = "." + w + "."
    for c1, c2 in zip(nw, nw[1:]):
        bigram = (c1, c2)
        count[bigram] = count.get(bigram, 0) + 1
```

```python
sorted(count.items(), key= lambda kv : -kv[1])
```

```python
chars = sorted(list(set("".join(words))))
stoi = {s: i + 1 for i, s in enumerate(chars)} # char -> index
stoi['.'] = 0
itos = {i:s for s,i in stoi.items()} # index -> char
print(stoi)
print(itos)
```

```python
for w in words:
    chs = '.' + w + '.'
    for ch1, ch2 in zip(w, w[1:]):
        ix1 = stoi[ch1]
        ix2 = stoi[ch2]
        N[ix1, ix2] += 1
N
```
