```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('./datasets/FuelConsumption.csv')
df.head() # prints default top 5 column
# df.tail() # prints default bottom 5 column
# df.info() # prints the info of the columns
# df.describe() # describes data all the details about column
```

```python
cdf = df[['ENGINESIZE', 'CYLINDERS', "FUELCONSUMPTION_COMB", "CO2EMISSIONS"]]
cdf.head(9)
```

```python
vis = df[['ENGINESIZE', 'CYLINDERS', "FUELCONSUMPTION_COMB", "CO2EMISSIONS"]]
vis.hist()
plt.show()
```

```python
# Plotting the graph to see which feature has great impact on co2 emission
# plot the vs graph with engine size vs co2 emission, cylinders vs emission, fuel consm vs emission
# Engine size vs emission

plt.scatter(cdf.FUELCONSUMPTION_COMB, cdf.CO2EMISSIONS)
plt.xlabel("Fuel Consumption")
plt.ylabel("Co2 Emission")
plt.show()
```

```python

plt.scatter(cdf.ENGINESIZE, cdf.CO2EMISSIONS)
plt.xlabel("Engine Size")
plt.ylabel("Co2 Emission")
plt.show()
```

```python
plt.scatter(cdf.CYLINDERS, cdf.CO2EMISSIONS) 
plt.xlabel("Cylinders")
plt.ylabel("Co2 Emissions")
plt.show()
```
